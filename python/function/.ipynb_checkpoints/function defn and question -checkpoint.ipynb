```python
# 13. Check palindrome number
def palindrome(n):
    original = n
    rev= 0 
    while n!=0:
        dig = n%10
        rev = rev*10+dig
        n = n//10

    if original == rev:
        return True
    else:
        return False
        
    
```

```python
n = int(input("enter any number"))
if palindrome(n):
    print(n,"palindrome number")
else:
    print(n,"not palindrome number")
```

```python
def second_largest(n):
    largest = float('-inf')
    second_largest = float('-inf')
    for i in n:
        if i>largest :
            second_largest = largest
            largest = i
        elif i>second_largest and i!=largest:
            second_largest = i

    return second_largest


second_largest([10,20,3,4,67])
```

```python
# frequency count .. 
def freq(s):
    dict = {}
    for i in s:
        if i in dict:
            dict[i] +=1
        else:
            dict[i] = 1
    return dict

freq("programming")
        
```

```python
# find missing number
def missing(nums,n):
    expected_sum = n*(n+1)//2
    actual_sum = sum(nums)
    return expected_sum - actual_sum
```

```python
missing([1,2,3,5],5)
```

```python
def missingNumber(nums,n):
    actual_sum = 0 
    expected_sum = n*(n+1)//2
    for i in nums:
        actual_sum += i
    return expected_sum - actual_sum
    
```

```python
# sort the string... 
def sort_string(s):
    l = list(s)
    
    for i in range(len(l)):
        for j in range(i+1,len(l)):
            if l[i]>l[j] :
                l[i],l[j] = l[j],l[i]
                
    sorted_s = "".join(l)

    return sorted_s

sort_string("programming")
```

```python
# 69. Print Pascal triangle (basic logic)
def pascal_triangle(n):
    prev_row = []

    for i in range(n):
        curr_row = []

        for j in range(i + 1):
            if j == 0 or j == i:
                curr_row.append(1)
            else:
                curr_row.append(prev_row[j - 1] + prev_row[j])

        for num in curr_row:
            print(num, end=" ")
        print()

        prev_row = curr_row


pascal_triangle(5)
```

```python
# 75. Print diamond pattern..
def diamond_pattern(n):
    for i in range(1,n+1):
        for j in range(n-i):
            print(" ",end = " ")
        for k in range(2*i-1):
            print("*",end = " ")
        print()

    for i in range(n-1,0,-1):
        for j in range(n-i):
            print(" ",end = " ")
        for k in range(2*i-1):
            print("*",end = " ")
        print()

diamond_pattern(5)
        
```

```python
# 76. Check Armstrong number.. 
def Armstrong_number(n):
    arm = n
    sum_arm = 0
    digit = len(str(n)) # count the number of digits that present in given number 
    while n>0:
        a = n%10
        b = a**digit
        sum_arm += b
        n = n//10
    if sum_arm == arm:
        print(arm,"is armstrong number")
    else:
        print(arm,"is not armstrong number: ")

Armstrong_number(9474)
    
```

```python
#77. Print Armstrong numbers till n
def arm_n(n):
    for i in range(1, n + 1):
        arm = i
        temp = i
        arm_sum = 0
        dig = len(str(i))

        while temp > 0:
            a = temp % 10
            arm_sum += a ** dig
            temp //= 10

        if arm == arm_sum:
            print(arm, end=" ")

arm_n(1000)
            
```

```python
# 78. Check perfect number.. 
def perfect_number(n):
    l = []
    sum = 0
    for i in range(1,n+1):
        if n%i==0:
            l.append(i)
    print(l)
    for j in range(len(l)-1):
        sum += l[j]
    print(sum)

    if sum == n:
        print(n,"is perfect number")
    else:
        print(n,"is not perfect number")
        

perfect_number(56)
```

```python
# 80. Check strong number.. 
def strng(n):
    org = n

    l = []
    while n>0:
        dig = n%10
        n = n//10
        fact = 1
        for i in range(1,dig+1):
            fact *= i
        l.append(fact)

        sum_l = 0
        for i in range(len(l)):
            sum_l += l[i]

    if sum_l == org:
        print(org,"is strong number")
    else:
        print(org,"not strong number")
        
strng(145)
        
```

```python
# 81. Print strong numbers till n..
def strong_number(n):

    for i in range(1, n + 1):
        org = i
        temp = i
        l = []

        while temp > 0:
            dig = temp % 10
            temp = temp // 10

            fact = 1
            for j in range(1, dig + 1):
                fact *= j

            l.append(fact)

        sum = 0
        for k in range(len(l)):
            sum += l[k]

        if sum == org:
            print(org,end = " ")

strong_number(1000)
```

```python
# 82. Find GCD of two numbers..(greatest common factor)
def Gcd(n,m):
    l1 = []
    l2 = []
    common = []
    for i in range(1,n+1):
        if n%i==0:
            l1.append(i)
    for j in range(1,m+1):
        if m%j==0:
            l2.append(j)

    for k in l1:
        if k in l2:
            common.append(k)
    print(common[-1])

Gcd(12,18)
        
```

```python
# 83. Find LCM of two numbers.. (least common multiplication)
def Lcm(a,b):
    l1 = []
    l2 = []
    common = []
    for i in range(1,a*b+1):
        c = i*a
        l1.append(c)
    for j in range(1,b*a+1):
        
        d = j*b
        l2.append(d)

    for k in l1:
        if k in l2:
            common.append(k)
    print(common[0])
            

Lcm(12,18)
        
```

```python
# 84. Check leap year..
def leap_year(n):

    if n%4 == 0:
        if n%100 ==0 :
            if n%400 == 0:
                print(n,"leap year")
            else:
                print(n,"not leap year")
        else:
            print(n,"leap year")
    else:
        print(n,'not leap year')

leap_year(2020)
```

```python
# 85. Print all leap years in range
def leap_year_range(start,stop):
    for i in range(start,stop+1):
        if i%4 == 0:
            if i%100 == 0:
                if i%400 == 0:
                    print(i,end = " ")
            else:
                print(i,end = " ")
        

leap_year_range(2010,2020)
```

```python
#92. Print digital root..
def digit_root(n):
    while n>=10:
        total = 0
        for i in str(n):
            total += int(i)

        n = total  
    print(n)
    
digit_root(2345)
```

```python
# 93. Count frequency of digits.. 
def digit_frequency(n):
    d = {}
    for i in str(n):
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
    print(d)

digit_frequency(2423432)
```

```python
from collections import Counter
def digit_frq(n):
    d =Counter(str(n))

    print(d)

digit_frq(23245353)
```

```python
# 94. Print unique digits... 
def unique_digit(n):
    l = []
    for i in str(n):
        if i not in l:
            l.append(i)
    print(l)

unique_digit(12356324)
            
        
```

```python
# 95. Print duplicate digits.
def duplicate_digit(n):
    l = []
    duplicate = []
    for i in str(n):
        if i not in l:
            l.append(i)
        else:
            duplicate.append(int(i))
    print(duplicate)
            

unique_digit(12356324)
```

```python
def duplicate(n):
    d = {}
    for i in str(n):
        if i in d:
            d[i] += 1
        else:
            d[i] = 1

    for key ,value in d.items():
        if value>1:
            print(key,end = " ")

duplicate(23433434442987)
```

```python
# 12. Reverse a number
def reverseNumber(n):
    rev = 0
    while n!= 0:
        dig = n%10
        rev = rev*10+dig
        n = n//10
    return rev
reverseNumber(123)
```

```python
# 13. Check palindrome number
def palindrome(n):
    original = n
    rev= 0 
    while n!=0:
        dig = n%10
        rev = rev*10+dig
        n = n//10

    if original == rev:
        return True
    else:
        return False
        
    
```

```python
n = int(input("enter any number"))
if palindrome(n):
    print(n,"palindrome number")
else:
    print(n,"not palindrome number")
```

```python
def second_largest(n):
    largest = float('-inf')
    second_largest = float('-inf')
    for i in n:
        if i>largest :
            second_largest = largest
            largest = i
        elif i>second_largest and i!=largest:
            second_largest = i

    return second_largest


second_largest([10,20,3,4,67])
```

```python
# frequency count .. 
def freq(s):
    dict = {}
    for i in s:
        if i in dict:
            dict[i] +=1
        else:
            dict[i] = 1
    return dict

freq("programming")
        
```

```python
s = input("enter any string")
freq(s)
```

```python
# find missing number
def missing(nums,n):
    expected_sum = n*(n+1)//2
    actual_sum = sum(nums)
    return expected_sum - actual_sum
```

```python
missing([1,2,3,5],5)
```

```python
def missingNumber(nums,n):
    actual_sum = 0 
    expected_sum = n*(n+1)//2
    for i in nums:
        actual_sum += i
    return expected_sum - actual_sum
    
```

```python
# sort the string... 
def sort_string(s):
    l = list(s)
    
    for i in range(len(l)):
        for j in range(i+1,len(l)):
            if l[i]>l[j] :
                l[i],l[j] = l[j],l[i]
                
    sorted_s = "".join(l)

    return sorted_s

sort_string("programming")
```

```python
# 47. Check if number is prime .. 
def prime_number(n):
    if n == 0:
        print(n,"not prime number")
    elif n == 1:
        print(n, "not prime number")
    else:
        for i in range(2,n):
            if n%i==0:
                print(n,"not prime number")
                break

        else:
            print(n,"prime number")
            
prime_number(12)
```

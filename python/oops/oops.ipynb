```python
class Atm:

    # constructor(special type of function) --> super power (without function call constructor call autometicallly 
    # when you create object.. )   
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input("""
        Hi how can i help you 
        1. press 1 to create pin
        2. press 2 to change pin 
        3. press 3 to check balance
        4. press 4 to withdraw
        5. anything else to exit
        """)

        if user_input == '1':
            # create pin 
            self.create_pin()
            
        elif user_input == '2':
            # change pin 
            self.change_pin()
            
        elif user_input == '3':
            # check balance
            self.check_balance()
            
        elif user_input == '4':
            # withdraw
            self.withdrawl()
        else: 
             exit()

            # created for generate pin 
    def create_pin(self):
        user_pin = input('enter your pin')
        self.pin = user_pin

        user_balance = int(input("enter your balance"))
        self.balance = user_balance

        print('pin created successfully: ')
        self.menu()

        # created for update pin 
    def change_pin(self):
        old_pin = input('enter your old pin')
        if old_pin == self.pin:
            new_pin = input('enter new pin: ')
            self.pin = new_pin
            print('pin changed sucessfull')
            self.menu()
        else:
            print("wrong old pin ")
            self.menu()

            # created for check balance
    def check_balance(self):
        user_password = input('enter your password: ')
        if user_password == self.pin:
            print(f'your account balance is: {self.balance}')
            self.menu()
        else:
            print('password is wrong: ')
            self.menu() 

    # created for withdrawl money 
    def withdrawl(self):
        user_pin = input('enter your password: ')
        if user_pin == self.pin:
            user_money = int(input('enter the amount :'))
            if user_money<=self.balance:
                self.balance = self.balance - user_money
                print(f'money withdraw successfull : {user_money}')
                print(f'current balance in your account : {self.balance}')
                self.menu()
            else:
                print('insufficient balance: ')
                self.menu()
        else:
            print("wrong password please enter correct password : ")
            self.menu()

```

```python
# create your own datatype.. 
class fraction:
    # parameterized constructor
    def __init__(self,x,y):
        self.num = x
        self.dnum = y

    def __str__(self):  # magic method
        return '{}/{}'.format(self.num,self.dnum)

    def __add__(self,other):
        new_num = self.num*other.dnum + other.num*self.dnum
        new_dnum = self.dnum*other.dnum

        return '{}/{}'.format(new_num,new_dnum)

    def __sub__(self,other):
        new_num = self.num*other.dnum - other.num*self.dnum
        new_dnum = self.dnum*other.dnum
    
        return '{}/{}'.format(new_num,new_dnum)
        
    def __mul__(self,other):
        new_num = self.num*other.num
        new_dnum = self.dnum*other.dnum

        return '{}/{}'.format(new_num,new_dnum)

    def __truediv__(self,other):
        new_num = self.num*other.dnum
        new_dnum = self.dnum*other.num

        return '{}/{}'.format(new_num,new_dnum)

    def convert_to_decimal(self):
        return self.num/self.dnum
        
        
```

```python
fr3 = fraction(3,4)
fr3.convert_to_decimal()
```

```python
fr1 = fraction(2,5)
fr2 = fraction(5,6)
```

```python
print(fr1)
print(fr2)
print(fr1+fr2)
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
```

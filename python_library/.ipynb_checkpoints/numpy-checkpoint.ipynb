```python
# numpy --> numpy stands for 'numerical python' powerful python library that is widely 
# used for scientific computing ,data , analysis ,and numerical computing task 

# numpy , an array is the foundamental object .. 

# array is used to store homogeneous data elements contiguous block of memory .. 
```

```python
import numpy as np
```

```python
l = [1,2,3,4,5]
arr = np.array(l)
print(arr)
print(arr.size)
print(arr.shape)
print(arr.ndim)
print(arr.dtype)
print(type(arr))
```

```python
arr = np.array([[10,20,30],[40,30,57]])

print(arr)
print(arr.size)
print(arr.shape)
print(arr.ndim)
print(arr.dtype)
print(type(arr))
```

```python
# 3d array
arr = np.array(
    [
        [
            [1,2,3],
            [4,5,6]
        ],
        [
            [4,5,6],
            [5,7,8]
        ]
    ]
)

print(arr)
print(arr.shape)
print(arr.ndim)
```

```python
arr = np.arange(7)
print(arr)
```

```python
arr = np.arange(1,9,1)
print(arr)

arr1 = arr.reshape(2,4)
print(arr1)
```

```python
arr = np.arange(1,28,1).reshape(3,3,3)
print(arr)
```

```python
arr = np.array([1,2,3],ndmin = 10)
print(arr)
```

```python
# arange() function -- always return 1d array
arr = np.arange(1,9)
print(arr)
```

```python
arr = np.arange(1,19).reshape(3,3,2)
print(arr)
```

```python
# element fetch 1d array.. 
arr = np.array([1,2,3,4,5])
print(arr[0])
print(arr[-1])
```

```python
arr = np.arange(1,13).reshape(4,3)
print(arr)
print(arr[3][2])
print()
print(arr[2])
```

```python
arr = np.arange(1,28).reshape(3,3,3)
print(arr)
print(arr.ndim)
```

```python
# direct iteration.
arr = np.array([10,20,30,40,50])
for i in arr:
    print(i,end = " ")
```

```python
# indirect iteration.. 
arr = np.array([10,20,30,40,50])
for i in range(len(arr)):
    print(arr[i],end = " ")
```

```python
arr = np.arange(1,13).reshape(3,4)
for i in arr:
    for j in i:
        print(j,end = " ")
```

```python
arr = np.arange(1,28).reshape(3,3,3)
for i in arr:
    for j in i:
        for k in j:
            print(k,end = " ")
```

```python
arr = np.arange(1,28).reshape(3,3,3)
for i in np.nditer(arr):
    print(i,end = " ")
```

```python
arr = np.arange(1,28).reshape(3,3,3)
for i,j in np.ndenumerate(arr):
    print(i,j)
```
